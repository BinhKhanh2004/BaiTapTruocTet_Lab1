### 1.1NUMPY

```python
import numpy as np
# Tạo một mảng 1 chiều
mang_1D= np.array([1,2,3,4,5,6,7,8])
print(mang_1D)
# Tạo một mảng 2 chiều
mang_2D= np.array(([1,3,5,7],[2,4,6,8]))
print(mang_2D)
# Tạo mảng ngẫu nhiên
mang_nD= np.random.rand(4,4)
print(mang_nD)
```

```python
# Tạo một mảng
mang_1D= np.array([1,2,3,4,5,6,7,8])
print(mang_1D)
# Cộng và nhân mảng với một số
print(mang_1D + 5)
print(mang_1D * 5)
# Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(mang_1D. sum())
print(mang_1D. mean())
print(mang_1D. max())
print(mang_1D. min())
```

```python
import pandas as pd

# Tạo dataframe và dictionary
data = {
    "Name": ["Khánh", "Hoàng", "Hân"],
    "Age": [2004, 2004, 2006],
    "Score": [10, 9, 10]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df["Name"])

# Lấy hàng thứ 2
print(df.iloc[1])

# Lấy người có tuổi lớn hơn 18
print(df[df["Age"]>18])

# tính điểm trung bình của người đó
print(df.iloc[2])
print(np.mean(df["Score"]))
```

```python
import matplotlib.pyplot as plt

x=[212,28,74,4,32,99]
y=[20,867,645,55,30,240]

plt.plot(x,y,label="Biểu đồ đường",color="pink",marker='o')
plt.title("Biểu đồ đường")
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend
plt.show
```

```python
Name = ["Khánh", "Tuấn", "Khương", "Hân", "Đức"]
Score = [8, 8.5, 9, 9.5, 10]

plt.bar(Name, Score, color = "red")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes=np.random.rand(100)*100
plt.scatter(x,y,s=sizes,alpha=0.5,c="blue")
plt.title("Biểu đồ phân tán")
plt.xlabel=("Trục x")
plt.ylabel=("Trục y")
plt.show()
```
